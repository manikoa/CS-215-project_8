```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

# Part 3: 

```python
time_sorted = pd.read_csv('time_sorted.csv')
search_hist = pd.read_csv('search_hist.csv')
```

```python
plt.bar(time_sorted['days'].value_counts().keys(), time_sorted['days'].value_counts()/len(time_sorted['days']), color = 'black', edgecolor = 'black')

plt.bar(search_hist['day'].value_counts().keys(), search_hist['day'].value_counts()/len(search_hist['day']), color = 'pink', edgecolor = 'black')

plt.xlabel("Day")
plt.ylabel("Percent of Responses")
```

We noticed that Maniko's data is much more evenly distributed amongst the different days. Jacob's data is very concentrated on certain days with larger gaps where no instances were found. It appears that most of Jacob's data can be found on the 11th and 22th days.
